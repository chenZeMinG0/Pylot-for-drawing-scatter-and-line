```python
"""Currently, this algorithm just can draw two kinds of graph: scatter chart and line chart
       else: Solve a problem where the graph can not display Chinese words.
             Abolish scientific notation for coordinate axes.
       
   Using the follow case to show the process
"""
# Python 列表操作的 大O 效率
```

```python
import timeit
import matplotlib.pyplot as plt
%matplotlib inline


#引用字体，让中文能够显示出来
from pylab import*
mpl.rcParams['font.sans-serif']=['SimHei']
```

```python
# 删除列表末尾所需的时间: pop()
aisx_x = []
aisx_y2 = []
for i in range(1, 11):
    x = list(range(2000000 * i))
    aisx_x.append((x[-1]+1))
    popend = timeit.Timer("x.pop()", "from __main__ import x")
    y1 = popend.timeit(number=1000)
    aisx_y2.append(y1)
```

```python
print(aisx_x)
print(aisx_y2)
```

```python
# 删除列表第一个所需的时间:  pop(0)
aisx_x = []
aisx_y1 = []
for i in range(1, 11):
    x = list(range(2000000 * i))
    aisx_x.append(x[-1] + 1)
    popzero = timeit.Timer("x.pop(0)", "from __main__ import x")
    y1 = popzero.timeit(number=1000)
    aisx_y1.append(y1)
    
```

```python
print(aisx_x)
print(aisx_y1)
```

# Standard codes of drawing scatter and line chart

```python
#将两组时间visualizing
#用图去展示执行pop()与pop(0) 所需要时间的对比


fig, ax = plt.subplots(figsize=(15, 9), dpi=128)


#取消坐标轴的科学显示法: 两种方法
#ax.get_xaxis().get_major_formatter().set_scientific(False)
plt.ticklabel_format(axis="x", style='plain')



#ax.scatter(aisx_x, aisx_y1)            #pop(0) 画点
ax.plot(aisx_x, aisx_y1, marker='o')     #画线,marker:线中有圆点。无论画线还是画点，marker都是共用的

#ax.scatter(aisx_x, aisx_y2, marker='v')  #pop() 画点 marker表示形状
ax.plot(aisx_x, aisx_y2, marker='v')     #画线，marker：线中有倒三角




# 添加标题和标签
ax.set_title("对比 pop(0) 与 pop() 的性能", fontsize=24)
ax.set_xlabel("列表长度", fontsize=15)
ax.set_ylabel("执行时间 （单位：毫秒）", fontsize=15)



#添加图例
'''
plt.legend(loc)中的loc取值(图例的位置)：
 0:'best'
 1: 'upper right'
 2: 'upper left'
 3:'lower left'
 4: 'lower right'
 5: 'right'
 6: 'center left'
 7: 'center right'
 8: 'lower center'
 9: 'upper center'
 10: 'center'
'''
plt.legend(loc=2, labels=['pop(0)', 'pop()'], fontsize=20)


```
